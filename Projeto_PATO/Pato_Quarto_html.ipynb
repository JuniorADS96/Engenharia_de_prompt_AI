```python
%%html
<!DOCTYPE html>
<html lang="pt-BR">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Pato Matemático - Treino de Cálculo Mental</title>
    <style>
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }

        body {
            font-family: 'Comic Sans MS', 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            background: linear-gradient(135deg, #90EE90 0%, #3CB371 100%);
            min-height: 100vh;
            display: flex;
            justify-content: center;
            align-items: center;
            padding: 20px;
        }

        .container {
            background: #E8F8E8;
            border-radius: 30px;
            box-shadow: 0 20px 60px rgba(0, 100, 0, 0.3);
            max-width: 600px;
            width: 100%;
            padding: 40px;
            text-align: center;
            border: 5px solid #66BB6A;
            position: relative;
        }

        .container::before {
            content: '✨';
            position: absolute;
            top: -15px;
            left: 20px;
            font-size: 2em;
            animation: twinkle 1.5s infinite;
        }

        .container::after {
            content: '✨';
            position: absolute;
            top: -15px;
            right: 20px;
            font-size: 2em;
            animation: twinkle 1.5s infinite 0.5s;
        }

        @keyframes twinkle {
            0%, 100% { opacity: 1; transform: scale(1); }
            50% { opacity: 0.5; transform: scale(0.8); }
        }

        /* Telas */
        .screen {
            display: none;
        }

        .screen.active {
            display: block;
            animation: fadeIn 0.3s ease-in;
        }

        @keyframes fadeIn {
            from {
                opacity: 0;
                transform: scale(0.95);
            }
            to {
                opacity: 1;
                transform: scale(1);
            }
        }

        /* Tela de Boas-vindas */
        .welcome-title {
            font-size: 3em;
            color: #2D6A2D;
            margin-bottom: 20px;
            font-weight: 900;
            letter-spacing: 2px;
            text-transform: uppercase;
            text-shadow: 3px 3px 0px rgba(102, 187, 106, 0.5);
        }

        .welcome-subtitle {
            font-size: 1.3em;
            color: #4CAF50;
            margin-bottom: 40px;
            font-style: italic;
            font-weight: 600;
        }

        .welcome-icon {
            font-size: 5em;
            margin-bottom: 20px;
            animation: bounce 2s infinite;
            display: inline-block;
        }

        @keyframes bounce {
            0%, 100% { transform: translateY(0) rotate(0deg); }
            50% { transform: translateY(-25px) rotate(5deg); }
        }

        /* Tela de Nome */
        .name-section h2 {
            font-size: 2.2em;
            color: #2D6A2D;
            margin-bottom: 30px;
            font-weight: 900;
            text-shadow: 2px 2px 0px rgba(102, 187, 106, 0.3);
        }

        .name-section p {
            color: #4CAF50;
            margin-bottom: 20px;
            font-size: 1.15em;
            font-weight: 600;
        }

        input[type="text"],
        input[type="number"] {
            width: 100%;
            padding: 15px;
            font-size: 1.1em;
            border: 3px solid #66BB6A;
            border-radius: 20px;
            margin-bottom: 20px;
            transition: all 0.3s;
            background: white;
            color: #2D6A2D;
            font-weight: 600;
        }

        input[type="text"]:focus,
        input[type="number"]:focus {
            outline: none;
            border-color: #2D6A2D;
            box-shadow: 0 0 20px rgba(45, 106, 45, 0.4);
            transform: scale(1.02);
        }

        input[type="text"]::placeholder,
        input[type="number"]::placeholder {
            color: #81C784;
        }

        /* Tela do Jogo */
        .game-header {
            margin-bottom: 30px;
        }

        .game-title {
            font-size: 2em;
            color: #2D6A2D;
            margin-bottom: 10px;
            text-transform: uppercase;
            font-weight: 900;
            text-shadow: 2px 2px 0px rgba(102, 187, 106, 0.3);
        }

        .game-info {
            display: flex;
            justify-content: space-around;
            margin-bottom: 20px;
            gap: 20px;
        }

        .info-item {
            flex: 1;
            background: linear-gradient(135deg, #C8E6C9 0%, #A5D6A7 100%);
            padding: 15px;
            border-radius: 15px;
            border: 3px solid #66BB6A;
            box-shadow: 0 5px 15px rgba(0, 100, 0, 0.2);
        }

        .info-label {
            font-size: 0.9em;
            color: #2D6A2D;
            margin-bottom: 8px;
            font-weight: 700;
        }

        .info-value {
            font-size: 1.6em;
            color: #1B5E20;
            font-weight: 900;
        }

        /* Personagem - SVG do Pato Fofo */
        .character-container {
            height: 280px;
            margin: 30px 0;
            display: flex;
            justify-content: center;
            align-items: center;
            position: relative;
            background: linear-gradient(135deg, #F1F8F1 0%, #E1F5E1 100%);
            border-radius: 25px;
            border: 3px dashed #81C784;
            box-shadow: inset 0 4px 15px rgba(0, 100, 0, 0.1);
        }

        .character-image {
            max-width: 100%;
            max-height: 100%;
            filter: drop-shadow(0 10px 25px rgba(0, 100, 0, 0.2));
        }

        /* Animações do Pato */
        @keyframes jump {
            0%, 100% { transform: translateY(0) rotate(0deg); }
            50% { transform: translateY(-35px) rotate(-5deg); }
        }

        @keyframes thinking {
            0%, 100% { transform: rotate(-3deg); }
            50% { transform: rotate(3deg); }
        }

        @keyframes celebrate {
            0%, 100% { transform: scale(1) rotate(0deg); }
            25% { transform: scale(1.15) rotate(-15deg); }
            50% { transform: scale(1.1) rotate(15deg); }
            75% { transform: scale(1.15) rotate(-10deg); }
        }

        .character-image.jump {
            animation: jump 0.8s ease-in-out;
        }

        .character-image.thinking {
            animation: thinking 1.2s ease-in-out infinite;
        }

        .character-image.celebrate {
            animation: celebrate 1s ease-in-out;
        }

        /* Problema Matemático */
        .math-problem {
            background: linear-gradient(135deg, #FFFFFF 0%, #F1F8F1 100%);
            border: 4px solid #66BB6A;
            border-radius: 20px;
            padding: 30px;
            margin: 30px 0;
            box-shadow: 0 8px 20px rgba(0, 100, 0, 0.15);
        }

        .math-expression {
            font-size: 2.8em;
            color: #2D6A2D;
            font-weight: 900;
            font-family: 'Arial', monospace;
            text-shadow: 2px 2px 0px rgba(102, 187, 106, 0.3);
        }

        /* Input de Resposta */
        .answer-section {
            margin: 30px 0;
        }

        .answer-label {
            display: block;
            color: #2D6A2D;
            margin-bottom: 15px;
            font-size: 1.1em;
            font-weight: 700;
        }

        .answer-input-container {
            position: relative;
            display: flex;
            gap: 10px;
        }

        #answerInput {
            flex: 1;
            padding: 18px;
            font-size: 1.4em;
            border: 3px solid #66BB6A;
            border-radius: 15px;
            text-align: center;
            transition: all 0.3s;
            background: white;
            color: #2D6A2D;
            font-weight: 700;
        }

        #answerInput:focus {
            outline: none;
            border-color: #2D6A2D;
            box-shadow: 0 0 20px rgba(45, 106, 45, 0.4);
            transform: scale(1.02);
        }

        #answerInput.error {
            border-color: #FF6B6B;
            background-color: #FFE4E4;
        }

        #answerInput.correct {
            border-color: #1B5E20;
            background-color: #C8E6C9;
        }

        /* Timer */
        .timer {
            font-size: 1.3em;
            color: #2D6A2D;
            font-weight: 800;
            margin-top: 15px;
        }

        .timer.warning {
            animation: pulse 0.5s infinite;
            color: #FF6B6B;
        }

        @keyframes pulse {
            0%, 100% { color: #FF6B6B; transform: scale(1); }
            50% { color: #FF9999; transform: scale(1.1); }
        }

        /* Botões */
        button {
            background: linear-gradient(135deg, #66BB6A 0%, #4CAF50 100%);
            color: white;
            border: none;
            padding: 18px 40px;
            font-size: 1.2em;
            border-radius: 15px;
            cursor: pointer;
            transition: all 0.2s;
            font-weight: 800;
            text-transform: uppercase;
            letter-spacing: 1.5px;
            box-shadow: 0 6px 15px rgba(0, 100, 0, 0.3);
        }

        button:hover {
            transform: translateY(-4px) scale(1.05);
            box-shadow: 0 10px 25px rgba(0, 100, 0, 0.4);
        }

        button:active {
            transform: translateY(-1px);
        }

        button:disabled {
            opacity: 0.5;
            cursor: not-allowed;
        }

        .button-container {
            display: flex;
            gap: 12px;
            justify-content: center;
            flex-wrap: wrap;
        }

        .button-secondary {
            background: linear-gradient(135deg, #81C784 0%, #66BB6A 100%);
        }

        /* Resultado */
        .result-message {
            font-size: 1.6em;
            font-weight: 900;
            margin: 25px 0;
            min-height: 50px;
            text-shadow: 1px 1px 3px rgba(0, 0, 0, 0.1);
        }

        .result-message.correct {
            color: #1B5E20;
            animation: slideIn 0.4s ease-out;
        }

        .result-message.incorrect {
            color: #FF6B6B;
            animation: shake 0.4s ease-in-out;
        }

        @keyframes slideIn {
            from { opacity: 0; transform: translateY(-20px); }
            to { opacity: 1; transform: translateY(0); }
        }

        @keyframes shake {
            0%, 100% { transform: translateX(0); }
            25% { transform: translateX(-10px); }
            75% { transform: translateX(10px); }
        }

        /* Estatísticas */
        .stats {
            background: linear-gradient(135deg, #C8E6C9 0%, #A5D6A7 100%);
            border-radius: 20px;
            padding: 25px;
            margin: 25px 0;
            border: 3px solid #66BB6A;
            box-shadow: 0 5px 15px rgba(0, 100, 0, 0.15);
        }

        .stat-row {
            display: flex;
            justify-content: space-between;
            margin: 12px 0;
            font-size: 1.2em;
        }

        .stat-label {
            color: #2D6A2D;
            font-weight: 700;
        }

        .stat-value {
            color: #1B5E20;
            font-weight: 900;
        }

        /* Responsivo */
        @media (max-width: 480px) {
            .container {
                padding: 25px;
            }

            .welcome-title {
                font-size: 2em;
            }

            .game-title {
                font-size: 1.5em;
            }

            .math-expression {
                font-size: 2.2em;
            }

            .character-container {
                height: 200px;
            }

            button {
                padding: 14px 25px;
                font-size: 1em;
            }
        }
    </style>
</head>
<body>
    <div class="container">
        <!-- Tela de Boas-vindas -->
        <div id="welcomeScreen" class="screen active">
            <div class="welcome-icon">🦆</div>
            <h1 class="welcome-title">🌿 Pato Matemático 🌿</h1>
            <p class="welcome-subtitle">Treino de Cálculo Mental - Aventura do Pato Explorador</p>
            <button onclick="goToNameScreen()">✨ OK, Começar! ✨</button>
        </div>

        <!-- Tela de Nome -->
        <div id="nameScreen" class="screen">
            <div class="name-section">
                <h2>🦆 Qual é o seu nome? 🦆</h2>
                <p>Digite seu nome para começar a aventura matemática com o Pato Explorador! 🎩✨</p>
                <input
                    type="text"
                    id="playerName"
                    placeholder="Seu nome aqui..."
                    onkeypress="handleNameKeypress(event)"
                    maxlength="50"
                >
                <button onclick="startGame()">🚀 Iniciar Jogo 🚀</button>
            </div>
        </div>

        <!-- Tela do Jogo -->
        <div id="gameScreen" class="screen">
            <div class="game-header">
                <h1 class="game-title">🎮 Jogo Aberto do <span id="playerNameDisplay"></span> 🎮</h1>
                <div class="game-info">
                    <div class="info-item">
                        <div class="info-label">📝 Questão</div>
                        <div class="info-value"><span id="questionCount">1</span>/10</div>
                    </div>
                    <div class="info-item">
                        <div class="info-label">✅ Acertos</div>
                        <div class="info-value"><span id="correctCount">0</span></div>
                    </div>
                    <div class="info-item">
                        <div class="info-label">⭐ Pontuação</div>
                        <div class="info-value"><span id="score">0</span></div>
                    </div>
                </div>
            </div>

            <div class="character-container">
                <svg id="characterImage" class="character-image" viewBox="0 0 240 280" width="180" height="220">
                    <!-- Sombra -->
                    <ellipse cx="120" cy="260" rx="70" ry="15" fill="#00000010"/>

                    <!-- Corpo fofo principal -->
                    <ellipse cx="120" cy="160" rx="55" ry="70" fill="#FFD700"/>
                    <ellipse cx="120" cy="155" rx="52" ry="67" fill="#FFED4E"/>

                    <!-- Peito branco fofo -->
                    <ellipse cx="120" cy="180" rx="40" ry="50" fill="#FFFACD"/>

                    <!-- Pescoço -->
                    <circle cx="120" cy="90" r="42" fill="#FFED4E"/>

                    <!-- Cabeça redonda e fofa -->
                    <circle cx="120" cy="45" r="50" fill="#FFED4E"/>
                    <circle cx="120" cy="40" r="48" fill="#FFE44D"/>

                    <!-- Topo da cabeça (mais escuro) -->
                    <ellipse cx="120" cy="15" rx="38" ry="25" fill="#FFD700"/>

                    <!-- Chapéu Indiana Jones fofo -->
                    <ellipse cx="120" cy="-5" rx="45" ry="18" fill="#8B6F47"/>
                    <path d="M 75 0 L 80 -20 Q 120 -35 160 -20 L 165 0" fill="#A0826D" stroke="#654321" stroke-width="2"/>
                    <path d="M 72 0 Q 120 22 168 0" fill="none" stroke="#654321" stroke-width="2"/>
                    <!-- Aba do chapéu -->
                    <path d="M 80 2 Q 120 30 160 2" fill="#8B6F47" stroke="#654321" stroke-width="2"/>
                    <!-- Fita do chapéu -->
                    <path d="M 82 5 Q 120 12 158 5" stroke="#D2691E" stroke-width="4" fill="none"/>

                    <!-- Olhos grandes e brilhantes -->
                    <!-- Olho esquerdo -->
                    <circle cx="95" cy="35" r="12" fill="white"/>
                    <circle cx="95" cy="35" r="9" fill="#000000"/>
                    <circle cx="98" cy="32" r="5" fill="white" opacity="0.9"/>
                    <circle cx="99" cy="33" r="2" fill="#FFD700" opacity="0.6"/>

                    <!-- Olho direito -->
                    <circle cx="145" cy="35" r="12" fill="white"/>
                    <circle cx="145" cy="35" r="9" fill="#000000"/>
                    <circle cx="148" cy="32" r="5" fill="white" opacity="0.9"/>
                    <circle cx="149" cy="33" r="2" fill="#FFD700" opacity="0.6"/>

                    <!-- Sombreado dos olhos (pálpebra fofa) -->
                    <path d="M 82 40 A 12 12 0 0 1 108 40" fill="#FFD700" opacity="0.4"/>
                    <path d="M 132 40 A 12 12 0 0 1 158 40" fill="#FFD700" opacity="0.4"/>

                    <!-- Bico laranja fofo -->
                    <path d="M 120 50 L 132 62 L 120 66 Z" fill="#FF8C00"/>
                    <path d="M 120 50 L 108 62 L 120 66 Z" fill="#FF6347"/>

                    <!-- Sorriso feliz -->
                    <path d="M 108 70 Q 120 82 132 70" stroke="#FF6347" stroke-width="3" fill="none" stroke-linecap="round"/>

                    <!-- Bochechas rosadas fofas -->
                    <circle cx="75" cy="50" r="15" fill="#FFB6C1" opacity="0.7"/>
                    <circle cx="165" cy="50" r="15" fill="#FFB6C1" opacity="0.7"/>

                    <!-- Asas/Braços fofinhos -->
                    <!-- Asa esquerda -->
                    <ellipse cx="70" cy="155" rx="25" ry="50" fill="#FFD700" transform="rotate(-35 70 155)"/>
                    <path d="M 60 130 Q 50 145 55 175" fill="none" stroke="#FFE44D" stroke-width="4" stroke-linecap="round" opacity="0.7"/>

                    <!-- Asa direita -->
                    <ellipse cx="170" cy="155" rx="25" ry="50" fill="#FFD700" transform="rotate(35 170 155)"/>
                    <path d="M 180 130 Q 190 145 185 175" fill="none" stroke="#FFE44D" stroke-width="4" stroke-linecap="round" opacity="0.7"/>

                    <!-- Patas dianteiras -->
                    <ellipse cx="95" cy="225" rx="18" ry="25" fill="#FFA500"/>
                    <ellipse cx="95" cy="250" rx="22" ry="18" fill="#FF8C00"/>

                    <ellipse cx="145" cy="225" rx="18" ry="25" fill="#FFA500"/>
                    <ellipse cx="145" cy="250" rx="22" ry="18" fill="#FF8C00"/>

                    <!-- Detalhes das patas -->
                    <ellipse cx="90" cy="252" rx="6" ry="4" fill="#FF6347"/>
                    <ellipse cx="100" cy="252" rx="6" ry="4" fill="#FF6347"/>
                    <ellipse cx="140" cy="252" rx="6" ry="4" fill="#FF6347"/>
                    <ellipse cx="150" cy="252" rx="6" ry="4" fill="#FF6347"/>

                    <!-- Cinto/Correia de couro -->
                    <rect x="100" y="185" width="40" height="15" rx="3" fill="#8B6F47"/>
                    <rect x="103" y="188" width="34" height="10" rx="2" fill="#A0826D"/>

                    <!-- Fivela do cinto -->
                    <rect x="113" y="180" width="14" height="20" fill="#DAA520"/>
                    <rect x="115" y="183" width="10" height="14" fill="#FFD700"/>

                    <!-- Bolsa/Mochila fofa -->
                    <ellipse cx="65" cy="155" rx="28" ry="50" fill="#8B6F47"/>
                    <ellipse cx="65" cy="155" rx="25" ry="46" fill="#A0826D"/>

                    <!-- Detalhes da bolsa -->
                    <circle cx="65" cy="140" r="7" fill="#654321"/>
                    <circle cx="65" cy="170" r="7" fill="#654321"/>
                    <rect x="55" y="185" width="20" height="8" rx="2" fill="#654321"/>

                    <!-- Chicote/Corda na mão -->
                    <path d="M 50 130 Q 25 110 15 75 Q 10 50 8 30" fill="none" stroke="#8B6F47" stroke-width="5" stroke-linecap="round"/>
                    <path d="M 50 130 Q 25 110 15 75 Q 10 50 8 30" fill="none" stroke="#A0826D" stroke-width="3" stroke-linecap="round"/>

                    <!-- Nó da corda -->
                    <circle cx="8" cy="30" r="6" fill="#8B6F47"/>
                    <circle cx="8" cy="30" r="4" fill="#A0826D"/>

                    <!-- Adaga/Faca no cinto -->
                    <rect x="128" y="182" width="5" height="28" fill="#C0C0C0" rx="1"/>
                    <polygon points="130.5,182 127,176 134,176" fill="#D3D3D3"/>
                    <rect x="124" y="208" width="13" height="7" fill="#8B6F47"/>

                    <!-- Brilho nos olhos (mais vida) -->
                    <circle cx="102" cy="28" r="5" fill="white" opacity="0.9"/>
                    <circle cx="152" cy="28" r="5" fill="white" opacity="0.9"/>
                </svg>
            </div>

            <div class="math-problem">
                <div class="math-expression" id="mathExpression">0 + 0</div>
            </div>

            <div class="answer-section">
                <label class="answer-label">🎯 Qual é o resultado?</label>
                <input
                    type="number"
                    id="answerInput"
                    placeholder="Sua resposta..."
                    onkeypress="handleAnswerKeypress(event)"
                    disabled
                >
                <div class="timer">
                    ⏱️ <span id="timerDisplay">15</span>s
                </div>
            </div>

            <div class="result-message" id="resultMessage"></div>

            <div class="button-container">
                <button id="submitBtn" onclick="submitAnswer()" disabled>✅ Enviar Resposta ✅</button>
                <button class="button-secondary" onclick="goToMenuScreen()">⚙️ Menu ⚙️</button>
            </div>

            <div class="stats">
                <div class="stat-row">
                    <span class="stat-label">⏱️ Tempo Médio:</span>
                    <span class="stat-value"><span id="avgTime">0</span>s</span>
                </div>
                <div class="stat-row">
                    <span class="stat-label">🎯 Taxa de Acerto:</span>
                    <span class="stat-value"><span id="accuracy">0</span>%</span>
                </div>
            </div>
        </div>

        <!-- Tela de Pausa/Menu -->
        <div id="menuScreen" class="screen">
            <h1 class="welcome-title">📋 Menu 📋</h1>
            <p style="color: #2D6A2D; font-size: 1.2em; margin: 20px 0; font-weight: 700;">Bem-vindo, <strong id="menuPlayerName"></strong>! 👋</p>
            <div style="margin: 30px 0;">
                <button onclick="resumeGame()" style="width: 100%; margin-bottom: 10px;">▶️ Continuar Jogo ▶️</button>
                <button class="button-secondary" onclick="resetGame()" style="width: 100%; margin-bottom: 10px;">🔄 Novo Jogo 🔄</button>
                <button class="button-secondary" onclick="goToWelcomeScreen()" style="width: 100%;">🏠 Início 🏠</button>
            </div>
        </div>

        <!-- Tela de Fim de Jogo -->
        <div id="endScreen" class="screen">
            <div class="welcome-icon">🏆✨🏆</div>
            <h1 class="welcome-title">🎉 Parabéns! 🎉</h1>
            <div class="stats">
                <div class="stat-row">
                    <span class="stat-label">🦆 Jogador:</span>
                    <span class="stat-value" id="endPlayerName"></span>
                </div>
                <div class="stat-row">
                    <span class="stat-label">✅ Questões Corretas:</span>
                    <span class="stat-value"><span id="endCorrect">0</span>/10</span>
                </div>
                <div class="stat-row">
                    <span class="stat-label">⭐ Pontuação Total:</span>
                    <span class="stat-value"><span id="endScore">0</span></span>
                </div>
                <div class="stat-row">
                    <span class="stat-label">🎯 Taxa de Acerto:</span>
                    <span class="stat-value"><span id="endAccuracy">0</span>%</span>
                </div>
                <div class="stat-row">
                    <span class="stat-label">⏱️ Tempo Médio:</span>
                    <span class="stat-value"><span id="endAvgTime">0</span>s</span>
                </div>
            </div>
            <div style="margin: 30px 0;">
                <button onclick="resetGame()" style="width: 100%; margin-bottom: 10px;">🎮 Jogar Novamente 🎮</button>
                <button class="button-secondary" onclick="goToWelcomeScreen()" style="width: 100%;">🏠 Início 🏠</button>
            </div>
        </div>
    </div>

    <script>
        // Estado do Jogo
        let gameState = {
            playerName: '',
            currentQuestion: 0,
            correctAnswers: 0,
            totalScore: 0,
            timeTaken: [],
            currentProblem: {},
            timeRemaining: 15,
            timerInterval: null,
            answered: false,
            paused: false
        };

        // Navegação de Telas
        function showScreen(screenId) {
            document.querySelectorAll('.screen').forEach(screen => {
                screen.classList.remove('active');
            });
            document.getElementById(screenId).classList.add('active');
        }

        function goToWelcomeScreen() {
            resetGameState();
            showScreen('welcomeScreen');
        }

        function goToNameScreen() {
            showScreen('nameScreen');
            setTimeout(() => document.getElementById('playerName').focus(), 100);
        }

        function goToMenuScreen() {
            gameState.paused = true;
            if (gameState.timerInterval) clearInterval(gameState.timerInterval);
            document.getElementById('menuPlayerName').textContent = gameState.playerName;
            showScreen('menuScreen');
        }

        function resumeGame() {
            gameState.paused = false;
            showScreen('gameScreen');
            startTimer();
        }

        // Entrada do Nome
        function handleNameKeypress(event) {
            if (event.key === 'Enter') {
                startGame();
            }
        }

        function startGame() {
            const name = document.getElementById('playerName').value.trim();
            if (name === '') {
                alert('Por favor, digite seu nome!');
                return;
            }
            gameState.playerName = name;
            document.getElementById('playerNameDisplay').textContent = name;
            showScreen('gameScreen');
            nextQuestion();
        }

        // Geração de Problemas Matemáticos
        function generateProblem() {
            const difficulty = Math.min(Math.floor(gameState.currentQuestion / 3) + 1, 3);
            const maxNum = 10 * difficulty;
            const operations = ['+', '-', '*'];
            const operation = operations[Math.floor(Math.random() * operations.length)];

            let num1 = Math.floor(Math.random() * maxNum) + 1;
            let num2 = Math.floor(Math.random() * maxNum) + 1;

            let result;
            if (operation === '+') {
                result = num1 + num2;
            } else if (operation === '-') {
                result = Math.abs(num1 - num2);
                if (num1 < num2) [num1, num2] = [num2, num1];
            } else {
                result = num1 * num2;
            }

            return {
                expression: `${num1} ${operation} ${num2}`,
                result: result
            };
        }

        // Próxima Questão
        function nextQuestion() {
            if (gameState.currentQuestion >= 10) {
                endGame();
                return;
            }

            gameState.currentQuestion++;
            gameState.currentProblem = generateProblem();
            gameState.timeRemaining = 15;
            gameState.answered = false;

            document.getElementById('questionCount').textContent = gameState.currentQuestion;
            document.getElementById('mathExpression').textContent = gameState.currentProblem.expression;
            document.getElementById('answerInput').value = '';
            document.getElementById('answerInput').className = '';
            document.getElementById('resultMessage').textContent = '';
            document.getElementById('answerInput').disabled = false;
            document.getElementById('submitBtn').disabled = false;

            animateCharacter('thinking');

            setTimeout(() => {
                document.getElementById('answerInput').focus();
                startTimer();
            }, 500);
        }

        // Timer
        function startTimer() {
            if (gameState.timerInterval) clearInterval(gameState.timerInterval);

            gameState.timerInterval = setInterval(() => {
                if (!gameState.answered && !gameState.paused) {
                    gameState.timeRemaining--;
                    document.getElementById('timerDisplay').textContent = gameState.timeRemaining;

                    if (gameState.timeRemaining <= 5) {
                        document.getElementById('timerDisplay').parentElement.classList.add('warning');
                    }

                    if (gameState.timeRemaining <= 0) {
                        clearInterval(gameState.timerInterval);
                        submitAnswer(true);
                    }
                }
            }, 1000);
        }

        // Enviar Resposta
        function handleAnswerKeypress(event) {
            if (event.key === 'Enter' && !gameState.answered) {
                submitAnswer();
            }
        }

        function submitAnswer(timeout = false) {
            if (gameState.answered) return;

            gameState.answered = true;
            clearInterval(gameState.timerInterval);
            document.getElementById('timerDisplay').parentElement.classList.remove('warning');

            const userAnswer = parseInt(document.getElementById('answerInput').value) || 0;
            const correctAnswer = gameState.currentProblem.result;
            const timeTaken = 15 - gameState.timeRemaining;

            gameState.timeTaken.push(timeTaken);

            if (userAnswer === correctAnswer) {
                gameState.correctAnswers++;
                gameState.totalScore += Math.max(100 - (timeTaken * 5), 10);

                document.getElementById('resultMessage').textContent = timeout ? '⏱️ Tempo esgotado, mas a resposta foi correta! 🎉' : '✅ Correto! Excelente! 🌟';
                document.getElementById('resultMessage').classList.add('correct');
                document.getElementById('answerInput').classList.add('correct');

                animateCharacter('celebrate');
            } else {
                document.getElementById('resultMessage').textContent = `❌ Incorreto! A resposta é ${correctAnswer}`;
                document.getElementById('resultMessage').classList.add('incorrect');
                document.getElementById('answerInput').classList.add('error');

                animateCharacter('sad');
            }

            document.getElementById('correctCount').textContent = gameState.correctAnswers;
            document.getElementById('score').textContent = gameState.totalScore;
            updateStats();

            document.getElementById('answerInput').disabled = true;
            document.getElementById('submitBtn').disabled = true;

            setTimeout(nextQuestion, 3000);
        }

        // Atualizar Estatísticas
        function updateStats() {
            const avgTime = gameState.timeTaken.length > 0
                ? (gameState.timeTaken.reduce((a, b) => a + b, 0) / gameState.timeTaken.length).toFixed(1)
                : 0;
            const accuracy = gameState.currentQuestion > 0
                ? Math.round((gameState.correctAnswers / gameState.currentQuestion) * 100)
                : 0;

            document.getElementById('avgTime').textContent = avgTime;
            document.getElementById('accuracy').textContent = accuracy;
        }

        // Fim do Jogo
        function endGame() {
            const accuracy = Math.round((gameState.correctAnswers / 10) * 100);
            const avgTime = (gameState.timeTaken.reduce((a, b) => a + b, 0) / gameState.timeTaken.length).toFixed(1);

            document.getElementById('endPlayerName').textContent = gameState.playerName;
            document.getElementById('endCorrect').textContent = gameState.correctAnswers;
            document.getElementById('endScore').textContent = gameState.totalScore;
            document.getElementById('endAccuracy').textContent = accuracy;
            document.getElementById('endAvgTime').textContent = avgTime;

            showScreen('endScreen');
        }

        // Resetar Jogo
        function resetGame() {
            resetGameState();
            goToNameScreen();
        }

        function resetGameState() {
            gameState = {
                playerName: '',
                currentQuestion: 0,
                correctAnswers: 0,
                totalScore: 0,
                timeTaken: [],
                currentProblem: {},
                timeRemaining: 15,
                timerInterval: null,
                answered: false,
                paused: false
            };
            clearInterval(gameState.timerInterval);
        }

        // Animações do Personagem
        function animateCharacter(pose) {
            const img = document.getElementById('characterImage');
            img.classList.remove('jump', 'thinking', 'celebrate');

            setTimeout(() => {
                if (pose === 'thinking') {
                    img.classList.add('thinking');
                } else if (pose === 'celebrate') {
                    img.classList.add('celebrate');
                } else if (pose === 'jump') {
                    img.classList.add('jump');
                }
            }, 10);
        }
    </script>
</body>
</html>
```
